```python
df3 = pd.read_csv('../../Datasets/Production_Data.csv')
df3
```

```python
workers = df3['Worker ID'].nunique()
print(f'{workers} trabalhadores identificados')
```

```python
rejeitados = df3['Qty Rejected'].sum()
print(f'{rejeitados} produtos rejeitados')
```

```python
ordens = df3['Case ID'].nunique()
print(f'{ordens} ordens de produção processadas')
```

```python
# Quero separar a data do tempo HH:MM:SS
df3['Start Data'] = df3['Start Timestamp'].str.split(' ').str[0]
df3['Start Time'] = df3['Start Timestamp'].str.split(' ').str[1]

df3['Complete Data'] = df3['Complete Timestamp'].str.split(' ').str[0]
df3['Complete Time'] = df3['Complete Timestamp'].str.split(' ').str[1]

df3.drop(columns = ['Start Timestamp', 'Complete Timestamp'], inplace = True)
```

```python
df3['Start Time'].min()
```

```python
print(f'Essas são as datas em que o processamento começa mais cedo às {df3["Start Time"].min()}:\n {df3[df3["Start Time"] == df3["Start Time"].min()]["Start Data"].unique()}')
```

```python
print(f'Essas são as datas em que o processamento começa mais tarde às {df3["Start Time"].max()}:\n {df3[df3["Start Time"] == df3["Start Time"].max()]["Start Data"].unique()}')
# df3[df3['Start Time'] == df3['Start Time'].max()]['Start Data'].unique()
```

```python
import datetime
df3["Start Data"] = pd.to_datetime(df3["Start Data"])
df3["Complete Data"] = pd.to_datetime(df3["Complete Data"])
dias_producao = df3["Complete Data"].max() - df3["Start Data"].min()
dias_producao.days
```

```python
rejeitado_por_dia = []
for dia in df3["Start Data"].unique():
    i += 1
    rejeitado = df3[df3["Start Data"] == dia]["Qty Rejected"].sum() / len(df3[df3["Start Data"] == dia]["Qty Rejected"])
    rejeitado_por_dia.append((rejeitado))

media_rejeitado = sum(rejeitado_por_dia) / len(rejeitado_por_dia)
media_rejeitado
```

```python
df3.columns
```

```python
op_por_dia = []
for dia in df3["Start Data"].unique():
    i += 1
    op = df3[df3["Start Data"] == dia]["Work Order  Qty"].sum() / len(df3[df3["Start Data"] == dia]["Work Order  Qty"])
    op_por_dia.append((op))

media_op = sum(op_por_dia) / len(op_por_dia)
media_op
```

```python
df3["Start Month"] = df3["Start Data"].dt.month
dic_produzidos = {}
for i in df3["Start Month"].unique():
    lista_produzidos = []
    df_mes = df3[df3["Start Month"] == i]
    for j in df_mes["Start Data"].unique():
        produzidos = df_mes[df_mes["Start Data"] == j]["Work Order  Qty"].sum()/len(df_mes[df_mes["Start Data"] == j]["Work Order  Qty"])
        print(f'No mês {i} dia {j} foram produzidos em média {produzidos} produtos')
        lista_produzidos.append(produzidos)
    dic_produzidos[i] = lista_produzidos
```

```python
for key, value in dic_produzidos.items():
    print(f'No mês {key} foram produzidos em média {sum(value)/len(value)} produtos')
```

```python
df3["Start Month"] = df3["Start Data"].dt.month
dic_rejeitados = {}
for i in df3["Start Month"].unique():
    lista_rejeitados = []
    df_mes = df3[df3["Start Month"] == i]
    for j in df_mes["Start Data"].unique():
        rejeitados = df_mes[df_mes["Start Data"] == j]["Qty Rejected"].sum()/len(df_mes[df_mes["Start Data"] == j]["Qty Rejected"])
        print(f'No mês {i} dia {j} foram rejeitados em média {rejeitados} produtos')
        lista_rejeitados.append(rejeitados)
    dic_rejeitados[i] = lista_rejeitados
```

```python
for key, value in dic_rejeitados.items():
    print(f'No mês {key} foram rejeitados em média {sum(value)/len(value)} produtos')
```

```python
import re
# Filtrando apenas o formato de horas HH:MM
df3["Span"] = df3["Span"].apply(lambda x: re.findall(r'\d{2}:\d{2}', x)[0])
df3["Span"].unique()

hours = df3["Span"].str.split(':').str[0].astype(float).mul(60)
minutes = df3["Span"].str.split(':').str[1].astype(float)
df3["Process Time"] = hours.add(minutes)
df3
```
